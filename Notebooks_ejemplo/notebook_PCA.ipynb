Uso típico 1 de una PCA

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

Generamos nube 2D elíptica y rotada

```python
rng = np.random.RandomState(0)
n = 600

# Varianzas muy distintas (elipse) + rotación (correlación)
sigma1, sigma2 = 3.0, 0.6              # "estiramiento" real
theta = np.deg2rad(35)                 # rotación

R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
S = np.diag([sigma1, sigma2])

X = rng.randn(n, 2) @ S @ R.T          # (n,2): elipse rotada
# Opcional: desplazar la nube para que se vea que PCA centra (pero luego StandardScaler centra)
X += np.array([2.0, -1.0])
```

Figura A: datos originales (x1,x2)

```python
plt.figure(figsize=(5.2, 4.2))
plt.scatter(X[:, 0], X[:, 1], s=12, alpha=0.6)
plt.xlabel("feature x1")
plt.ylabel("feature x2")
plt.title("Nube original en (x1, x2): elipse rotada (correlación)")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

PCA a 2D (aquí ya estamos en 2D, pero la gracia es el cambio de base)

```python
pipe = Pipeline([
    ("scaler", StandardScaler()),       # centra y escala (opcional en 2D, pero típico en clase)
    ("pca", PCA(n_components=2, random_state=0))
])
Z = pipe.fit_transform(X)               # coordenadas en la base de PCs

pca = pipe.named_steps["pca"]
r = pca.explained_variance_ratio_
print(f"Explained variance ratio: PC1={r[0]:.3f}, PC2={r[1]:.3f}")
```

Figura B: datos en coordenadas PCA (z1,z2)

```python
plt.figure(figsize=(5.2, 4.2))
plt.scatter(Z[:, 0], Z[:, 1], s=12, alpha=0.6)
plt.xlabel("PC1 (z1)")
plt.ylabel("PC2 (z2)")
plt.title("Tras PCA: nube alineada y 'estirada' en PC1")
plt.axis("equal")
plt.tight_layout()
plt.show()
```
